```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_csv('pes2013A.txt', sep = " ")
```

```python
dados.describe()
```

```python
dados.shape
```

```python
n=dados.shape[0]
```

```python
dados['V0302'] = dados['V0302'].astype('category')
dados.V0302.cat.categories = ('Masculino', 'Feminino')

#-------------

dados['V0404'] = dados['V0404'].astype('category')      
dados.V0404.cat.categories = ('Indígena', 'Branca', 'Preta', 'Amarela', 'Parda', 'Sem Declaração')

#------------

dados['V6002'] = dados['V6002'].astype('category')
dados.V6002.cat.categories = ('Publico', 'Privado', 'NA')

#-----------

dados['V4802'] = dados['V4802'].astype('category')
dados.V4802.cat.categories = ('Maternal', 'Alfabetização', 'Alfabetização de adultos', '1ª a 4ª séries',
                                '5ª a 8ª séries', '1ª série', '2ª a 5ª séries', '6ª a 9ª séries',
                                'Não seriado do ensino fundamental', 'Ensino médio', 'Superior', 'Não Aplicavel')

#---------- Sabe ler e escrever ------

dados['V0601'] = dados['V0601'].astype('category')
dados.V0601.cat.categories = ('Sim', 'Não')

#-----------Horas Trabalhadas/Semana -----

dados['V4707'] = dados['V4707'].astype('category')
dados.V4707.cat.categories = ('Até 14 horas','15 a 39 horas','40 a 44 horas','45 a 48 horas','49 horas ou mais','NA')

dados['UF'] = dados['UF'].astype('category')
dados.UF.cat.categories = ("Rondonia", "Acre", "Amazonas", "Roraima", "Pará","Amapá",
                                "Tocantins", "Maranhão", "Piauí", "Ceará", "Rio Grande do Norte",
                                "Paraíba", "Pernambuco", "Alagoas", "Sergipe", "Bahia", "Minas Gerais",
                                "Espírito Santo", "Rio de Janeiro", "São Paulo", "Paraná", "Santa Catarina",
                                "Rio Grande do Sul", "Mato Grosso do Sul", "Mato Grosso", "Goiás", "Distrito Federal")


```

```python
dados.V0302 = dados.V0302.astype('category')
dados.V0404 = dados.V0404.astype('category')
dados.v6002 = dados.V6002.astype('category')
dados.v4802 = dados.V4802.astype('category')
dados.v0601 = dados.V0601.astype('category')
dados.V4707 = dados.V4707.astype('category') 
```

```python
print('Frequências relativas por Sexo:')
s1 = (dados.V0302.value_counts()/n*100).round(decimals=1)
print(s1,'\n')

#-------------------------

print('Raça:')
r1 = (dados.V0404.value_counts()/n*100).round(decimals=1)
print(r1,'\n')

#------------------------

print('Frequências relativas por Instituição:')
i1= (dados.V6002.value_counts()/n*100)
print(i1)

#-----------------------

print("""
Grau de ensino:""")
f1= (dados.V4802.value_counts()/n*100)
print(f1)

#------------------------

print("""
Sabe ler e escrever""")
l1= (dados.V0601.value_counts()/n*100)
print(l1)

#------------------------

print("""
Horas Trabalhadas/Semana""")
t1= (dados.V4707.value_counts()/n*100)
print(t1)
```

```python
plot = s1.plot(kind='bar', title = 'Sexo', figsize=(6, 6),color=('r','b'))
plt.ylabel('Freq.Relativa (%)')
plt.xlabel('Gênero')

#------------------
```

```python
plot = s1.plot(kind='pie', title = 'Sexo',autopct='%.1f', figsize=(6, 6))
```

```python
Final = pd.crosstab(dados.UF, dados.V0302)
Final
```

```python
plot = r1.plot(kind='bar',title='Raça',figsize=(6, 6),color=('b','g','r','c','m', 'y'))
plt.ylabel('Freq.Relativa (%)')
plt.xlabel('Raça')
```

```python
plot = r1.plot(kind='pie',title='Raça',autopct='%.1f',figsize=(6, 6))
```

```python
Final = pd.crosstab(dados.UF, dados.V0404)
Final
```

```python
plot = i1.plot(kind='bar', title = 'Instituição', figsize=(6, 6),color=('b','g','r'))
plt.ylabel('Freq.Relativa (%)')
plt.xlabel('Rede de Ensino')
```

```python
plot = i1.plot(kind='pie',title='Rede de Ensino',autopct='%.1f',figsize=(6, 6))
```

```python
Final = pd.crosstab(dados.UF, dados.V6002)
Final
```

```python
plot = f1.plot(kind='bar',title='Grau de ensino',figsize=(6, 6),color=('b','g','r','c','m','y'))
plt.ylabel('Freq. Relativa (%)')
plt.xlabel('Grau de ensino')
```
